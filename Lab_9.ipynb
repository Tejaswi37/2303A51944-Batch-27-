```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/archive (11).zip')
```

```python
print("Dataset shape:", data.shape)
print("\nColumns:\n", data.columns)
print("\nMissing values:\n", data.isnull().sum())
```

```python
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality']
wine_df = data[features]
```

```python
sns.set(style="ticks", palette="coolwarm")

pair_plot = sns.pairplot(
    wine_df,
    hue='quality',        # Color code by wine quality
    diag_kind='kde',      # Use KDE plots on the diagonal
    corner=True           # Avoid duplicate plots
)
```

```python
pair_plot.fig.suptitle("Pair Plot: Chemical Properties vs Wine Quality", y=1.02)
plt.show()
```
