```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Create NumPy array for 12 monthly temperatures (°C)
temps_celsius = np.array([15.2, 16.5, 20.1, 25.3, 28.4, 32.5, 33.8, 33.2, 30.1, 26.3, 21.0, 17.5])

months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun","Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
```

```python
# Step 2: Calculate average temperature
avg_temp = np.mean(temps_celsius)

# Step 3: Find hottest and coldest months
hottest_month_index = np.argmax(temps_celsius)
coldest_month_index = np.argmin(temps_celsius)

hottest_month = months[hottest_month_index]
coldest_month = months[coldest_month_index]
```

```python
# Step 4: Normalize the temperature values (0–1 scale)
normalized_temps = (temps_celsius - temps_celsius.min()) / (temps_celsius.max() - temps_celsius.min())

# Step 5: Identify months above annual average
months_above_avg = months[temps_celsius > avg_temp]
temps_above_avg = temps_celsius[temps_celsius > avg_temp]

# Step 6: Convert Celsius to Fahrenheit
temps_fahrenheit = (temps_celsius * 9/5) + 32
# Step 7: Display results
print("Monthly Temperatures (°C):", temps_celsius)
print("Average Temperature (°C):", round(avg_temp, 2))
print(f"Hottest Month: {hottest_month} ({temps_celsius[hottest_month_index]}°C)")
print(f"Coldest Month: {coldest_month} ({temps_celsius[coldest_month_index]}°C)")
print("Normalized Temperatures:", np.round(normalized_temps, 2))
print("Months Above Average:", months_above_avg)
print("Temps in Fahrenheit:", np.round(temps_fahrenheit, 2))
```

```python
# Step 8 (Optional): Visualize temperature change
plt.figure(figsize=(10, 5))
plt.plot(months, temps_celsius, marker='o', linestyle='-', color='orange', label="Temperature (°C)")
plt.axhline(y=avg_temp, color='blue', linestyle='--', label="Average Temp")
plt.title("Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)
plt.show()
```
