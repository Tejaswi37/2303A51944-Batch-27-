```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    "Subject": ["Math", "Science", "English", "Math", "Science", "English", "Math", "Science", "English"],
    "Score": [88, 92, 85, 75, 89, 80, 90, 95, 78],
    "Student": ["A", "A", "A", "B", "B", "B", "C", "C", "C"]
}

df = pd.DataFrame(data)


```

```python
sns.set_theme(style="darkgrid")
```

```python
plot = sns.catplot(
    data=df,
    x="Subject",
    y="Score",
    kind="box",
    palette="viridis"
)
```

```python
plot.fig.suptitle("Exam Score Distribution by Subject", fontsize=14)
plot.set_axis_labels("Subject", "Exam Score")
plt.grid(True, linestyle="--", alpha=0.6)  # extra grid enhancement

plt.show()
```
