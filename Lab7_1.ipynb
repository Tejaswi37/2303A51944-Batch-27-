```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv('/content/archive (13).zip')
```

```python
print("Dataset Preview:")
print(data.head())
```

```python
data.columns = ['Country_Code', 'Country_Name', 'Year', 'CO2_Emissions']
```

```python
data['Year'] = data['Year'].astype(int)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['CO2_Emissions'], color='teal', linewidth=2.5, marker='o')
```

```python
plt.title('Global CO Emissions Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO Emissions (Million Metric Tons)', fontsize=12)
```

```python
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
```

```python
plt.show()
```
