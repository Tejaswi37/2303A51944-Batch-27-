```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
data = {
    'StudyTime': np.random.normal(15, 5, 50).clip(5, 30),      # hours/week
    'Attendance': np.random.normal(85, 10, 50).clip(60, 100),  # %
    'ExamScore': np.random.normal(70, 15, 50).clip(40, 100),   # marks
    'Gender': np.random.choice(['Male', 'Female'], 50)
}

df = pd.DataFrame(data)
```

```python
print(df.head())
```

```python
sns.set(style="whitegrid", palette="Set2")

pair_plot = sns.pairplot(
    df,
    hue='Gender',
    diag_kind='kde',
    corner=True
)
```

```python
pair_plot.fig.suptitle("Pair Plot: Student Performance Trends", y=1.02)
plt.show()
```
