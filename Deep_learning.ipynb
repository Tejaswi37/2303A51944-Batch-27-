```python
"""
Deep Learning-based Clustering for Financial Health Segmentation
Corrected / robust version
"""
import os
import random
import argparse
from datetime import datetime

import numpy as np
import pandas as pd

# --- ML / DL ---
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.neighbors import NearestNeighbors

import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, backend as K

# Reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)


# ------------------------ Utilities ------------------------

def load_dataset(path):
    try:
        df = pd.read_csv(path)
        print(f"✅ Loaded dataset: {df.shape[0]} rows, {df.shape[1]} cols")
        return df
    except FileNotFoundError as e:
        raise FileNotFoundError(f"Could not find file at: {path}") from e
    except Exception as e:
        raise RuntimeError(f"Error loading dataset: {e}") from e


def select_features(df):
    features = [
        'monthly_income_usd',
        'monthly_expenses_usd',
        'savings_usd',
        'debt_to_income_ratio',
        'savings_to_income_ratio'
    ]
    missing = [f for f in features if f not in df.columns]
    if missing:
        raise ValueError(f"Missing expected columns in dataset: {missing}\nAvailable columns: {list(df.columns)}")
    return df[features].copy()


def preprocess_features(X_df, outlier_removal=True):
    """
    Returns:
      X_clean_df  : (pd.DataFrame) cleaned (no NA, outliers removed if chosen)
      X_scaled    : (np.ndarray) standardized features (StandardScaler)
      scaler      : scaler object for inverse transform if needed
    """
    X = X_df.copy()
    # Drop NA rows
    X = X.dropna().reset_index(drop=True)

    if outlier_removal:
        # remove rows that have any feature outside the (Q1 - 1.5*IQR, Q3 + 1.5*IQR) range
        Q1 = X.quantile(0.25)
        Q3 = X.quantile(0.75)
        IQR = Q3 - Q1
        mask = ~((X < (Q1 - 1.5 * IQR)) | (X > (Q3 + 1.5 * IQR))).any(axis=1)
        X = X.loc[mask].reset_index(drop=True)
        print(f"After outlier removal: {X.shape[0]} rows remain.")

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X.values.astype(np.float32))
    return X, X_scaled, scaler


def safe_evaluate_clusters(X, labels, name='clusters'):
    """Evaluate clusters but guard against single-cluster issues."""
    labels = np.array(labels)
    unique_labels = np.unique(labels)
    result = {'silhouette': np.nan, 'davies_bouldin': np.nan, 'calinski_harabasz': np.nan}

    if X.shape[0] == 0:
        print(f"Warning: No samples to evaluate for {name}.")
        return result

    if unique_labels.shape[0] < 2:
        print(f"Warning: Only {unique_labels.shape[0]} cluster(s) for {name}; skipping some metrics.")
        # silhouette and others require >=2 clusters; still try calinski_harabasz if possible
        try:
            ch = calinski_harabasz_score(X, labels)
            result['calinski_harabasz'] = ch
        except Exception:
            pass
        return result

    try:
        s = silhouette_score(X, labels)
        db = davies_bouldin_score(X, labels)
        ch = calinski_harabasz_score(X, labels)
        print(f"=== Evaluation for {name} ===")
        print(f"Silhouette Score: {s:.4f}")
        print(f"Davies-Bouldin Index: {db:.4f} (lower is better)")
        print(f"Calinski-Harabasz Index: {ch:.4f} (higher is better)")
        result = {'silhouette': float(s), 'davies_bouldin': float(db), 'calinski_harabasz': float(ch)}
    except Exception as e:
        print(f"Error computing cluster metrics for {name}: {e}")
    return result


# ------------------------ Baseline: KMeans ------------------------

def run_kmeans(X, n_clusters=4, random_state=SEED):
    X_np = np.asarray(X)
    if X_np.shape[0] < n_clusters:
        raise ValueError("Number of samples smaller than n_clusters for KMeans.")
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
    labels = kmeans.fit_predict(X_np)
    return kmeans, labels


# ------------------------ Autoencoder ------------------------

def build_autoencoder(input_dim, latent_dim=3):
    inputs = layers.Input(shape=(input_dim,))
    x = layers.Dense(64, activation='relu')(inputs)
    x = layers.Dense(32, activation='relu')(x)
    latent = layers.Dense(latent_dim, activation='relu', name='latent')(x)

    x = layers.Dense(32, activation='relu')(latent)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(input_dim, activation='linear')(x)

    ae = models.Model(inputs, outputs, name='autoencoder')
    encoder = models.Model(inputs, latent, name='encoder')

    ae.compile(optimizer=optimizers.Adam(learning_rate=1e-3), loss='mse')
    return ae, encoder


# ------------------------ VAE (minimal safe implementation) ------------------------

class VAE(tf.keras.Model):
    def __init__(self, input_dim, latent_dim=3):
        super(VAE, self).__init__()
        self.latent_dim = latent_dim
        # Encoder
        self.encoder_net = tf.keras.Sequential([
            layers.InputLayer(input_shape=(input_dim,)),
            layers.Dense(64, activation='relu'),
            layers.Dense(32, activation='relu'),
            layers.Dense(latent_dim * 2),  # will split into mean and logvar
        ])
        # Decoder
        self.decoder_net = tf.keras.Sequential([
            layers.InputLayer(input_shape=(latent_dim,)),
            layers.Dense(32, activation='relu'),
            layers.Dense(64, activation='relu'),
            layers.Dense(input_dim, activation='linear'),
        ])

    def encode(self, x):
        x = self.encoder_net(x)
        z_mean = x[:, :self.latent_dim]
        z_log_var = x[:, self.latent_dim:]
        return z_mean, z_log_var

    def reparameterize(self, mean, logvar):
        eps = tf.random.normal(shape=tf.shape(mean))
        return mean + tf.exp(0.5 * logvar) * eps

    def decode(self, z):
        return self.decoder_net(z)

    def call(self, inputs, training=False):
        z_mean, z_log_var = self.encode(inputs)
        z = self.reparameterize(z_mean, z_log_var)
        recon = self.decode(z)
        # compute losses and add them
        recon_loss = tf.reduce_mean(tf.reduce_sum(tf.square(inputs - recon), axis=1))
        kl_loss = -0.5 * tf.reduce_mean(tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=1))
        self.add_loss(recon_loss + kl_loss)
        return recon

    def get_z_mean(self, x):
        z_mean, _ = self.encode(x)
        return z_mean


# ------------------------ Simple DEC (approx) ------------------------

def target_distribution(q):
    # q: soft assignments (n_samples, n_clusters)
    q = np.array(q, dtype=np.float64)
    weight = (q ** 2) / (q.sum(axis=0) + 1e-12)
    p = (weight.T / (weight.sum(axis=1) + 1e-12)).T
    return p


def build_dec(input_dim, latent_dim=10, n_clusters=4):
    inputs = layers.Input(shape=(input_dim,))
    x = layers.Dense(500, activation='relu')(inputs)
    x = layers.Dense(500, activation='relu')(x)
    x = layers.Dense(2000, activation='relu')(x)
    latent = layers.Dense(latent_dim, name='latent')(x)
    clustering_layer = layers.Dense(n_clusters, activation='softmax', name='clustering')(latent)
    model = models.Model(inputs, clustering_layer)
    model.compile(optimizer=optimizers.Adam(1e-4), loss='kld')
    return model


# ------------------------ SOM (lightweight) ------------------------

def train_som(X, map_size=(10, 10), num_iterations=1000):
    X_np = np.asarray(X, dtype=np.float32)
    n_rows, n_cols = map_size
    dim = X_np.shape[1]
    if X_np.shape[0] < n_rows * n_cols:
        # replicate some samples if dataset is small
        choices = np.random.choice(range(X_np.shape[0]), size=n_rows * n_cols, replace=True)
    else:
        choices = np.random.choice(range(X_np.shape[0]), size=n_rows * n_cols, replace=False)

    weights = X_np[choices].reshape(n_rows, n_cols, dim).copy()

    def winner_node(x):
        dists = np.linalg.norm(weights - x.reshape(1, 1, dim), axis=2)
        return np.unravel_index(np.argmin(dists), (n_rows, n_cols))

    initial_lr = 0.1
    for t in range(num_iterations):
        i = np.random.randint(0, X_np.shape[0])
        x = X_np[i]
        bmu = winner_node(x)
        lr = initial_lr * (1 - t / float(num_iterations))
        radius = max(n_rows, n_cols) / 2.0 * (1 - t / float(num_iterations))
        for r in range(n_rows):
            for c in range(n_cols):
                dist_to_bmu = np.sqrt((r - bmu[0]) ** 2 + (c - bmu[1]) ** 2)
                if dist_to_bmu <= radius and radius > 0:
                    influence = np.exp(-dist_to_bmu ** 2 / (2.0 * (radius ** 2)))
                    weights[r, c, :] += lr * influence * (x - weights[r, c, :])

    flat_weights = weights.reshape(-1, dim)
    nbrs = NearestNeighbors(n_neighbors=1).fit(flat_weights)
    _, indices = nbrs.kneighbors(X_np)
    labels = indices.flatten()
    return labels, weights


# ------------------------ Visualization ------------------------

def visualize_2d(X_embedded, labels, title='2D projection'):
    try:
        plt.figure(figsize=(8, 6))
        # convert labels to strings for seaborn hue robustness
        labels_str = [str(int(l)) for l in labels]
        sns.scatterplot(x=X_embedded[:, 0], y=X_embedded[:, 1], hue=labels_str, palette='tab10', s=30, legend='full')
        plt.title(title)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Visualization skipped due to error: {e}")


# ------------------------ Main pipeline ------------------------

def main(args):
    # load dataset
    df = load_dataset(args.input)
    X_raw = select_features(df)

    # preprocess
    X_clean_df, X_scaled, scaler = preprocess_features(X_raw, outlier_removal=True)
    print("Preprocessing done. Samples:", X_scaled.shape[0])

    results = pd.DataFrame({'user_index': X_clean_df.index})

    # PCA for visualization baseline
    pca_2 = PCA(n_components=2, random_state=SEED)
    X_pca2 = pca_2.fit_transform(X_scaled)

    # Baseline KMeans
    try:
        print("\n-- Baseline KMeans --")
        kmeans, labels_k = run_kmeans(X_scaled, n_clusters=args.n_clusters)
        eval_k = safe_evaluate_clusters(X_scaled, labels_k, name='KMeans (raw features)')
        visualize_2d(X_pca2, labels_k, title='KMeans clusters (PCA2D)')
        results['kmeans'] = labels_k
    except Exception as e:
        print(f"KMeans failed: {e}")
        eval_k = {'silhouette': np.nan, 'davies_bouldin': np.nan, 'calinski_harabasz': np.nan}
        results['kmeans'] = -1

    # Autoencoder + KMeans in latent space
    eval_ae = {'silhouette': np.nan, 'davies_bouldin': np.nan, 'calinski_harabasz': np.nan}
    labels_ae = np.array([-1] * X_scaled.shape[0])
    try:
        print("\n-- Autoencoder + KMeans --")
        ae, encoder = build_autoencoder(input_dim=X_scaled.shape[1], latent_dim=args.ae_latent)
        ae.fit(X_scaled, X_scaled, epochs=args.epochs, batch_size=args.batch_size, validation_split=0.1, verbose=1)
        Z_ae = encoder.predict(X_scaled)
        _, labels_ae = run_kmeans(Z_ae, n_clusters=args.n_clusters)
        eval_ae = safe_evaluate_clusters(Z_ae, labels_ae, name='AE latent + KMeans')
        # visualize
        z_pca = PCA(n_components=2, random_state=SEED).fit_transform(Z_ae) if Z_ae.shape[1] > 2 else Z_ae
        visualize_2d(z_pca, labels_ae, title='AE latent clusters')
        results['ae_kmeans'] = labels_ae
    except Exception as e:
        print(f"Autoencoder step failed: {e}")
        results['ae_kmeans'] = labels_ae

    # VAE + KMeans
    eval_vae = {'silhouette': np.nan, 'davies_bouldin': np.nan, 'calinski_harabasz': np.nan}
    labels_vae = np.array([-1] * X_scaled.shape[0])
    try:
        print("\n-- VAE + KMeans --")
        vae_model = VAE(input_dim=X_scaled.shape[1], latent_dim=args.vae_latent)
        vae_model.compile(optimizer=optimizers.Adam(1e-3))
        # fit uses x both as input and target; VAE.add_loss handles loss in call
        vae_model.fit(X_scaled, X_scaled, epochs=args.epochs, batch_size=args.batch_size, validation_split=0.1, verbose=1)
        Z_vae = vae_model.get_z_mean(X_scaled)
        if isinstance(Z_vae, tf.Tensor):
            Z_vae = Z_vae.numpy()
        _, labels_vae = run_kmeans(Z_vae, n_clusters=args.n_clusters)
        eval_vae = safe_evaluate_clusters(Z_vae, labels_vae, name='VAE latent + KMeans')
        z_pca = PCA(n_components=2, random_state=SEED).fit_transform(Z_vae) if Z_vae.shape[1] > 2 else Z_vae
        visualize_2d(z_pca, labels_vae, title='VAE latent clusters')
        results['vae_kmeans'] = labels_vae
    except Exception as e:
        print(f"VAE step failed: {e}")
        results['vae_kmeans'] = labels_vae

    # DEC (approx) - simplified approach
    eval_dec = {'silhouette': np.nan, 'davies_bouldin': np.nan, 'calinski_harabasz': np.nan}
    labels_dec = np.array([-1] * X_scaled.shape[0])
    try:
        print("\n-- DEC (approx) --")
        dec_model = build_dec(input_dim=X_scaled.shape[1], latent_dim=args.dec_latent, n_clusters=args.n_clusters)
        # initialize training by predicting q and iterating
        q = dec_model.predict(X_scaled)
        p = target_distribution(q)
        for ite in range(args.dec_maxiter):
            dec_model.fit(X_scaled, p, epochs=1, batch_size=args.batch_size, verbose=0)
            q = dec_model.predict(X_scaled)
            p = target_distribution(q)
            if (ite + 1) % 10 == 0:
                print(f"DEC iter {ite+1}")
        labels_dec = q.argmax(axis=1)
        eval_dec = safe_evaluate_clusters(X_scaled, labels_dec, name='DEC (approx)')
        visualize_2d(X_pca2, labels_dec, title='DEC clusters (PCA 2D)')
        results['dec'] = labels_dec
    except Exception as e:
        print(f"DEC failed: {e}")
        results['dec'] = labels_dec

    # SOM
    eval_som = {'silhouette': np.nan, 'davies_bouldin': np.nan, 'calinski_harabasz': np.nan}
    labels_som = np.array([-1] * X_scaled.shape[0])
    try:
        print("\n-- SOM (lightweight) --")
        labels_som, som_weights = train_som(X_scaled, map_size=(8, 8), num_iterations=2000)
        eval_som = safe_evaluate_clusters(X_scaled, labels_som, name='SOM mapping labels')
        visualize_2d(X_pca2, labels_som, title='SOM clusters (PCA 2D)')
        results['som'] = labels_som
    except Exception as e:
        print(f"SOM failed: {e}")
        results['som'] = labels_som

    # ---------------- Summary & saving ----------------
    out_path = os.path.join(os.getcwd(), 'clustering_results.csv')
    try:
        results.to_csv(out_path, index=False)
        print(f"Saved clustering assignments to: {out_path}")
    except Exception as e:
        print(f"Could not save results CSV: {e}")

    # Compose evaluation summary and save
    eval_table = pd.DataFrame({
        'method': ['kmeans', 'ae_kmeans', 'vae_kmeans', 'dec', 'som'],
        'silhouette': [eval_k.get('silhouette'), eval_ae.get('silhouette'), eval_vae.get('silhouette'), eval_dec.get('silhouette'), eval_som.get('silhouette')],
        'davies_bouldin': [eval_k.get('davies_bouldin'), eval_ae.get('davies_bouldin'), eval_vae.get('davies_bouldin'), eval_dec.get('davies_bouldin'), eval_som.get('davies_bouldin')],
        'calinski_harabasz': [eval_k.get('calinski_harabasz'), eval_ae.get('calinski_harabasz'), eval_vae.get('calinski_harabasz'), eval_dec.get('calinski_harabasz'), eval_som.get('calinski_harabasz')]
    })
    print("\n=== Evaluation summary ===")
    print(eval_table)
    try:
        eval_table.to_csv('clustering_evaluation_summary.csv', index=False)
        print("Saved evaluation summary to clustering_evaluation_summary.csv")
    except Exception as e:
        print(f"Could not save evaluation summary: {e}")


if __name__ == '__main__':
    parser = argparse.ArgumentParser(description='Deep clustering pipeline for financial dataset')
    parser.add_argument('--input', type=str, default='/content/synthetic_personal_finance_dataset (1) (1).csv', help='Path to CSV dataset')
    parser.add_argument('--n_clusters', type=int, default=4, help='Number of clusters')
    parser.add_argument('--epochs', type=int, default=30, help='Training epochs for AE/VAE')
    parser.add_argument('--batch_size', type=int, default=256, help='Batch size for training')
    parser.add_argument('--ae_latent', type=int, default=3, help='Autoencoder latent dim')
    parser.add_argument('--vae_latent', type=int, default=3, help='VAE latent dim')
    parser.add_argument('--dec_latent', type=int, default=10, help='DEC latent dim')
    parser.add_argument('--dec_maxiter', type=int, default=30, help='DEC alternating iterations')
    args = parser.parse_args([]) # Pass an empty list to parse_args()

    main(args)
```
